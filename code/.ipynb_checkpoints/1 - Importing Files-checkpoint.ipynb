# Hierarchical bayesian estimation of psychometric parameters

```python
import pymc3 as pm
from theano import tensor as tt
```

In this section, the parameters of the psychometrics functions are estimated using bayesian modeling approache explicitely accounting for a contaminat parametre using the method describer in (Lee & Wagenmakers, 2013, Chap. 12). Regression parameters for the Intero and extero conditions are estimated both for slope and threshold. Here, we used a cummulative normal distribution to model the distribution.


- Lee, M. D., & Wagenmakers, E.-J. (2013). Bayesian cognitive modeling: A practical course. Cambridge University Press. https://doi.org/10.1017/CBO9781139087759

```python
def phi(x):
    # Cumulative distribution function for the standard normal distribution
    return (1+pm.math.erf(x/pm.math.sqrt(2)))/2

def find_nearest(array, value):
    idx = (np.abs(array-value)).argmin()
    return idx
```

```python
data = pd.DataFrame([])
for i, sub in enumerate(subList):
    df = pd.read_csv(path + f'/1_VPN_aux/{sub}/{sub[4:]}HRD/{sub[4:]}HRD_final.txt')
    for ii, modality in enumerate(['Extero', 'Intero']):
        this_df = df[df.Modality == modality]
        for intensity in np.arange(-40.5, 41, 0.5):
            # Store results
            data = data.append({'Subjects': i,
                                'Intensities': intensity,
                                'Modality': modality,
                                'Corrects': sum((this_df.Alpha == intensity) & (this_df.Estimation == 'More')),
                                'Total': sum(this_df.Alpha == intensity)}, ignore_index=True)
    
    idx = data[data.Modality == modality].Subjects.to_numpy()
data.Subjects = data.Subjects.astype('int')
```

```python
nSubjs = data.Subject.nunique()
x = data.Intensities.to_numpy()
n = data.Total.to_numpy()
r = data.Corrects.to_numpy()
idx = data.Subjects.to_numpy()
```

```python
with pm.Model() as model:

    # Variance priors
    sigma_a = pm.Uniform('sigma_a', lower=0, upper=100)
    sigma_b = pm.Uniform('sigma_b', lower=0, upper=100)

    # Mean priors
    mu_a = pm.Normal('mu_a', mu=0, tau=.001)
    mu_b = pm.Normal('mu_b', mu=0, tau=.001)

    alpha = pm.Normal('alpha', mu=mu_a, sd=sigma_a, shape=nSubjs)  # Threshold
    beta = pm.Normal('beta', mu=mu_b, sd=sigma_b, shape=nSubjs)  # Slope

    linerpredi = alpha[idx] + beta[idx] * data[data.Modality == modality]['Intensities'].to_numpy()
    thetaij = pm.Deterministic('thetaij', phi(linerpredi))

    rij_ = pm.Binomial('rij', p=thetaij, n=data[data.Modality == modality]['Total'].to_numpy(), 
                       observed=data[data.Modality == modality]['Corrects'].to_numpy())

    trace = pm.sample(init="advi+adapt_diag", tune=2000)
```

```python
sns.set_context('talk')
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
for i, modality, color in zip([0, 1], ['Intero', 'Extero'], ['firebrick', 'steelblue']):
    for sub in group_df[group_df.Session == 'Del2'].Subject.unique():
        try:
            x=group_df[(group_df.Session == 'Del1') & (group_df.Subject == sub) & (group_df.Modality == modality)].Threshold.iloc[0]
            y=group_df[(group_df.Session == 'Del2') & (group_df.Subject == sub) & (group_df.Modality == modality)].Threshold.iloc[0]
            axs[i].scatter(x, y, alpha=.2, color=color)
        except:
            pass
    axs[i].plot([-40, 40], [-40, 40], 'grey', linestyle='--')
    axs[i].set_xlim(-40, 40)
    axs[i].set_ylim(-40, 40)
    axs[i].set_xlabel('Del1')
    axs[i].set_ylabel('Del2')
    axs[i].set_title(modality)
sns.despine()
plt.tight_layout()
plt.savefig('testretest.png', dpi=600)
```
